```python
# legendas e anotações
# - colocando legendas;
# - nomeando os eixos;
# - escrevendo nos gráficos.

# - nomeando os eixos:
# - .xlabel(texto)
# - .ylabel(texto)
# parâmetro --> texto = nome daquele eixo

# - adicionando uma legenda:
# - .legend([Patch])
# parâmetro --> [patch] = uma lista de objetos Patch


# - para destacar alguma parte do gráfico:
# - .annotate(xy, xytexto, arrowprops)
# parâmetro --> xy = localização que queremos anotar
# parâmetro --> xttexto = localização do texto que queremos
# parâmetro --> arrowprops = desenha uma seta entre o texto e o ponto que queremos destacar.
```

```python
# criando títulos nos eixos
carros = pd.read_csv(caminho)
# renomeação das colunas
carros.columns = ['nomes', 'npg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']

npg = carros[['npg']]

x = range(1,10) # lista que varia de 1 até 10
y = [1, 2, 3, 4, .5, 4, 3, 2, 1]
plt.bar(x, y)
_= plt.xlabel('nosso título do eixo X')
_= plt.ylabel('nosso título do eixo Y')


```

```python
# criando legenda no gráfico 
import matplotlib.patches as npatches

# primeiro irei montar o gráfico
plt.plot(npg)
plt.xlabel('nomes dos carros')
plt.ylabel('milhas/galão')
plt.title('Consumo de gasolina') # serve para dar um título ao gráfico
plt.xticks(npg.index, carros.nomes, rotation=45) # para mostrar os nomes dos carros no eixo x
# agora irei criar a legenda
legenda = npatches.Patch(label='npg') # curva que será desenhada
_= plt.legend(handles=[legenda])
```

```python
# montando histogramas, box plots e scatter plots

caminho = '/content/mtcars.csv'

carros = pd.read_csv(caminho)
# renomeação das colunas
carros.columns = ['nomes', 'npg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
carros.index = carros.nomes
npg = carros['npg']

plt.hist(npg) # histograma com o plt
plt.show() # esconde o resultado da célula
```

```python
# histograma com seaborn
sns.distplot(npg) # histograma com uma linha que mostra a distribuição dos dados.
```

```python
# vendo os scatterplots em ação

carros.plot(kind='scatter', x='hp', y='npg', c=['darkgray'], s=150)
# scatterplots com pandas
```

```python
# scatterplots com seaborn
sns.regplot(x='hp', y='npg', data=carros, scatter=True)
```

```python
# gerando uma matriz de scatter plots
sns.pairplot(carros) # compara todas as variáveis com todas variáveis. - demora um pouco -
```

```python
# escolhendo algumas variáveis para análise
colunas = ['npg', 'disp', 'hp', 'wt']
novo_df = pd.DataFrame((carros[colunas].values), columns=colunas) # primeiro uso os colchetes para escolher as colunas desejadas..
# ..depois uso o atributo values para pegar apenas os valores dessas colunas..
# ..em seguida uso o parâmetro columns e passo para ele o objeto colunas para dar nomes as novas colunas do meu dataframe..
#..além disso criarei uma nova coluna onde irei definir os grupos dos meus carros.
carros_target = carros['am'].values
novo_df['grupo'] = pd.Series(carros_target, dtype='category') # separei os valores da coluna "am" do dataset original..
#.. e criei uma coluna chamada grupo no meu novo dataframe

# agora usarei a função pairplot do seaborn para criar uma nova matriz de scatter plot.
sns.pairplot(novo_df, hue='grupo', palette='hls') # também irei usar o hue para definir quais grupos irei colorir e quais as cores
```

```python
# construindo boxplots

carros.boxplot(column='npg', by='am')
carros.boxplot(column='wt', by='am')
```

```python
sns.boxplot(x='am', y='npg', data=carros, palette='hls')
```
