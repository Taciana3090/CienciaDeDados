```python
# agora iremos ver a transformação dos dados..
# para um formato onde torna-se mais fácil a análise.
# - removendo dados: elimando variáveis ou observações;
# - adicionando dados: acrescentando variáveis ou observações;
# - ordenando dados

import pandas as pd
import numpy as np

df = DataFrame(np.random.rand(36).reshape(6,6))
df



```

```python
# - removendo dados: elimando variáveis ou observações;
df.drop([0,2])
# deste modo foram removidas as linhas 0 e 2
```

```python
# - removendo dados: elimando variáveis ou observações;
df.drop([1,2], axis=1)
# deste modo foram removidas as colunas 1 e 2
```

```python
# - adicionando dados: acrescentando variáveis ou observações;
serie = pd.Series(np.arange(6))
serie.name = "nova_variavel"
serie
```

```python
# união dos dois dataframes
novo_df = pd.DataFrame.join(df, serie)
novo_df
```

```python
# - ordenando dados
df_ordenado = novo_df.sort_values(by=["nova_variavel"], ascending=[False])
df_ordenado
# foi ordenado pela coluna "nova_variavel" de maneira descrecente 
```
